```python
import seaborn as sns
from seaborn import lmplot
import sqlite3
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import re
import matplotlib as mpl
```

```python

pd.set_option('display.max_colwidth', None)

df=pd.read_csv('Web_access_log-akumenius.com.csv',names=["columna_unica"])
df.head(30)
```

```python
patro = ("[^\"]*")
columnes= df["columna_unica"].apply(lambda columna: re.findall(patro, columna))
df["web"]=columnes.map(lambda columna: columna[0].split(' ')[0])
df.head(30)
```

```python
df["ip"]=columnes.map(lambda columna: columna[0].split(' ')[1])
df.head(30)
```

```python
df["data_hora"]=columnes.map(lambda columna: columna[0].split(' ')[4 & 5])
df.head(30)
```

```python
df["get"]=columnes.map(lambda columna: columna[2])
df.head(30)
```

```python
df["numeros"]=columnes.map(lambda columna: columna[4])
df.head(30)
```

```python
df["pagina"]=columnes.map(lambda columna: columna[6])
df.head(30)
```

```python
df["navegador"]=columnes.map(lambda columna: columna[10] if len(columna)>=11 else "")
df.head(30)
```

```python
df=df.drop("columna_unica", axis=1)
df.head(30)
```
